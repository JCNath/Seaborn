```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
#sns.pairplot(df, hue="species")
# histogram
sns.histplot(data=df, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

```python
# Kernel Density Estimation
sns.kdeplot(data=df, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

```python
# Figure-level plots
sns.displot(data=df, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
```

```python
sns.displot(data=df, x="flipper_length_mm", hue="species", multiple="stack",kind='kde')
plt.show()
```

```python
sns.displot(data=df, x="flipper_length_mm", hue="species", col="species")
plt.show()
```

```python
tips = sns.load_dataset("tips")
g = sns.relplot(data=tips, x="total_bill", y="tip")
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
plt.show()
```

```python
g = sns.relplot(data=df, x="flipper_length_mm", y="bill_length_mm", col="sex")
g.set_axis_labels("Flipper length (mm)", "Bill length (mm)")
plt.show()
```
